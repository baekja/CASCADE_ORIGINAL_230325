```python
from rdkit import Chem
from collections import Counter
from tqdm import tqdm
tqdm.pandas()

smiles= pd.Series(data_rf.SMILES.unique())

def get_elements(smiles):
    mol = Chem.MolFromSmiles(smiles)
    elems = Counter((atom.GetSymbol() for atom in Chem.AddHs(mol).GetAtoms()))
    return pd.Series(elems)

elems = smiles.progress_apply(get_elements).fillna(0).astype(int)
```

```python
Y = np.concatenate(data_rf_final[['exp_result']].astype(int).to_numpy())
```

```python
# with grid search
from sklearn.metrics import accuracy_score, roc_auc_score, f1_score
from sklearn.metrics import classification_report,confusion_matrix

import numpy
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV

def split_data(X, Y,random_state, partition):

    from sklearn.model_selection import train_test_split

    x1, x2, y1, y2 = train_test_split(X, Y,test_size = partition, random_state=random_state)

    return x1, x2, y1, y2

def scan_parameters(x,y):


    n_estimators = [100,200,300,400]
    max_features = [1,2]
    max_depth = [15, 10, 100, 1000]

    params_dict = {'n_estimators': n_estimators, 'max_features' : max_features, 'max_depth': max_depth}

    scaler = StandardScaler(copy=True, with_mean=True, with_std=True)
    scaler.fit(x)
    x = scaler.transform(x)

    grid_search = GridSearchCV(estimator = RandomForestClassifier(), param_grid = params_dict, cv = 3,  n_jobs = 12)
    grid_search.fit(x, y)
#     print(grid_search.best_params_)

    return grid_search.best_params_

def train_RF(X_train, y_train, X_valid, y_valid, param_list = None):

    x_train = X_train
    x_valid = X_valid

    scaler = StandardScaler(copy=True, with_mean=True, with_std=True)
    scaler.fit(x_train)

    x_train = scaler.transform(x_train)

    if param_list == None:
        rf_classifier = RandomForestClassifier(n_estimators = 300)

    else:
        rf_classifier = RandomForestClassifier(n_estimators=param_list['n_estimators'], max_features = param_list['max_features'], max_depth = param_list['max_depth'])

    rf_classifier.fit(x_train, y_train)


    x_valid = scaler.transform(x_valid)
    predictions = rf_classifier.predict(x_valid)
    accuracy = (100*accuracy_score(y_valid, predictions))

    return accuracy

def build_RF_classifier(X,Y, option, model_name = None):

    if option == 'default':

        accuracies_default = []
        for i in range(10):
            x_train, x_valid, y_train, y_valid = split_data (X, Y,i,0.20)
            accuracies_default.append(train_RF(x_train, y_train, x_valid, y_valid))

        print('Average accuracy over 10 default runs: %.2f' % numpy.mean(accuracies_default))

    elif option == 'train':

        print('*-----------------------------*')
        print('Searchig for best parameters.')

        params = []
        accuracies = []

        for i in range(10):
            x_train, x_valid, y_train, y_valid = split_data (X, Y,i,0.20)
            best_parameters = scan_parameters(x_train, y_train)
            params.append(best_parameters)
            accuracy = train_RF(x_train, y_train, x_valid, y_valid, best_parameters)
            accuracies.append(accuracy)

        print('*-----------------------------*')
        print('Summary of Results.')
        print('*-----------------------------*')


        for i in range(len(accuracies)):
            print('Run ' + str (i+1)+ ' ', params[i], ' : ', accuracies[i])

        print('*-----------------------------*')
        print('Average of accuracy.')
        print('*-----------------------------*')
        print(numpy.average(accuracies))
```
